```python
df_plot = df.copy()
df_plot['taux_d’accès'] = pd.to_numeric(df_plot['taux_d’accès'], errors='coerce')
df_plot['capacite_formation'] = pd.to_numeric(df_plot['capacite_formation'], errors='coerce')
df_plot['effectif_total_candidats'] = pd.to_numeric(df_plot['effectif_total_candidats'], errors='coerce')
```

```python
top_formations = (
    df_plot.groupby('type_formation')['effectif_total_candidats']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_formations.values, y=top_formations.index)
for bar in ax.patches:
    ax.annotate(f"{int(bar.get_width())}",
                (bar.get_width(), bar.get_y() + bar.get_height()/2),
                va='center', ha='left', xytext=(5,0), textcoords='offset points')
plt.title("Top 10 des formations les plus demandées")
plt.xlabel("Nombre de candidats")
plt.tight_layout()
plt.show()
```

```python
type_counts = df_plot['type_formation'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(type_counts.values,
        labels=type_counts.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Répartition des types de formation")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df_plot, x='type_formation', y='taux_d’accès')
plt.xticks(rotation=45, ha='right')
plt.title("Distribution du taux d'accès par type de formation")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df_plot,
    x='capacite_formation',
    y='effectif_total_candidats',
    hue='type_formation',
    alpha=0.7
)
plt.title("Capacité vs nombre de candidats")
plt.xlabel("Capacité")
plt.ylabel("Candidats")
plt.tight_layout()
plt.show()
```

```python
top_regions = df_plot['region'].value_counts().head(10)

plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_regions.values, y=top_regions.index)
for bar in ax.patches:
    ax.annotate(f"{int(bar.get_width())}",
                (bar.get_width(), bar.get_y() + bar.get_height()/2),
                va='center', ha='left', xytext=(5,0), textcoords='offset points')
plt.title("Top 10 des régions")
plt.xlabel("Nombre de formations")
plt.tight_layout()
plt.show()
```

```python
pivot = pd.crosstab(df_plot['region'], df_plot['type_formation'])

plt.figure(figsize=(12,8))
sns.heatmap(pivot, annot=True, fmt="d", cmap="coolwarm")
plt.title("Répartition formations par région et type")
plt.tight_layout()
plt.show()
```
