**Formations les plus demandées**

```python
top_formations = (
    df_plot.groupby('type_formation')['effectif_total_candidats']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

sns.set_theme(style="whitegrid")

palette = sns.color_palette("viridis", len(top_formations))

plt.figure(figsize=(11,6))
ax = sns.barplot(x=top_formations.values, y=top_formations.index, palette=palette)

max_val = top_formations.values.max()
plt.xlim(0, max_val * 1.15)

for bar in ax.patches:
    width = bar.get_width()
    ax.annotate(
        f"{int(width):,}".replace(",", " "),
        (width, bar.get_y() + bar.get_height()/2),
        va='center',
        ha='left',
        xytext=(8, 0),
        textcoords='offset points',
        fontsize=9,
        fontweight='bold'
    )

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.title("Top 10 du type de formation les plus demandées", fontsize=14, weight='bold')
plt.xlabel("Nombre de candidats")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

Les licences sont le type de formation le plus demandée, on retrouve ensuite les BTS, les IFSI et les BUT.

**Etude de la sélectivité et des taux d'accès des formations**

```python
agg = (
    df_plot
    .groupby('type_formation')
    .agg({
        'capacite_formation': 'sum',
        'effectif_total_candidats': 'sum'
    })
    .reset_index()
)

agg_melt = agg.melt(id_vars='type_formation', 
                    value_vars=['capacite_formation', 'effectif_total_candidats'],
                    var_name='variable',
                    value_name='valeur')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12,6))
sns.lineplot(
    data=agg_melt,
    x='type_formation',
    y='valeur',
    hue='variable',
    marker='o'
)

for _, row in agg_melt.iterrows():
    plt.text(
        row.name % len(agg),  
        row['valeur'],
        f"{int(row['valeur'])}",
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.xticks(rotation=45, ha='right')
plt.title("Capacité vs Candidats par type de formation")
plt.xlabel("Type de formation")
plt.ylabel("Nombre")
plt.tight_layout()
plt.show()
```

**Analyse des candidatures et admissions par genre**

```python
df_plot = df.copy()

cols = [
    'effectif_total_candidats',
    'nombre_candidates',
    
]
df_plot[cols] = df_plot[cols].apply(pd.to_numeric, errors='coerce')


df_plot['candidates'] = df_plot['nombre_candidates'].fillna(0)
df_plot['candidats_hommes'] = df_plot['effectif_total_candidats'] - df_plot['candidates']


agg = pd.DataFrame({
    'Femmes': [
        df_plot['candidates'].sum(),
            ],
    'Hommes': [
        df_plot['candidats_hommes'].sum(),
            ]
}, index=['Candidats'])

sns.set_theme(style="whitegrid")

ax = agg.plot(kind='bar', figsize=(8,6), colormap='Set2')

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height,
            f"{int(height):,}".replace(",", " "),
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title("Nombre de candidats: Hommes VS Femmes")
plt.ylabel("Nombre")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

On remarque une grosse différence du nombre de candidats entre les femmes et les hommes.   
Pour 7 millions de femmes candidates sur ParcourSup, il y a seulement 5 millions d'hommes.  
Nous pouvons théoriser que les hommes s'ils poursuivent leurs études le font dans des formations spécialisées hors ParcourSup (CAP, BP...).

```python
df_plot = df.copy()

cols = [
    'effectif_total_candidats',
    'nombre_candidates',
    'total_admis_phase_principale',
    'total_admis_phase_complementaire',
    'total_candidates_ayant_accepte'
]
df_plot[cols] = df_plot[cols].apply(pd.to_numeric, errors='coerce')

df_plot['total_admis'] = (
    df_plot['total_admis_phase_principale'].fillna(0) +
    df_plot['total_admis_phase_complementaire'].fillna(0)
)

df_plot['candidates'] = df_plot['nombre_candidates'].fillna(0)
df_plot['candidats_hommes'] = df_plot['effectif_total_candidats'] - df_plot['candidates']

df_plot['admises'] = df_plot['total_candidates_ayant_accepte'].fillna(0)
df_plot['admis_hommes'] = df_plot['total_admis'] - df_plot['admises']

agg = (
    df_plot
    .groupby('type_formation')[[
        'candidates','candidats_hommes','admises','admis_hommes'
    ]]
    .sum()
    .reset_index()
)

agg['pct_femmes_candidats'] = agg['candidates'] / (agg['candidates'] + agg['candidats_hommes']) * 100
agg['pct_hommes_candidats'] = agg['candidats_hommes'] / (agg['candidates'] + agg['candidats_hommes']) * 100

agg['pct_femmes_admis'] = agg['admises'] / (agg['admises'] + agg['admis_hommes']) * 100
agg['pct_hommes_admis'] = agg['admis_hommes'] / (agg['admises'] + agg['admis_hommes']) * 100

agg_melt = agg.melt(
    id_vars='type_formation',
    value_vars=[
        'pct_femmes_candidats','pct_hommes_candidats',
        'pct_femmes_admis','pct_hommes_admis'
    ],
    var_name='categorie',
    value_name='pourcentage'
)

palette = {
    'pct_femmes_candidats': '#f4a6a6',
    'pct_hommes_candidats': '#6baed6',
    'pct_femmes_admis': '#e74c3c',
    'pct_hommes_admis': '#2171b5'
}

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=agg_melt,
    x='type_formation',
    y='pourcentage',
    hue='categorie',
    palette=palette
)

for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.text(
            bar.get_x() + bar.get_width()/2,
            height,
            f"{round(height):.0f}%",
            ha='center',
            va='bottom',
            fontsize=8
        )

plt.xticks(rotation=45, ha='right')
plt.title("Répartition Hommes / Femmes par formation (Candidats vs Admis)")
plt.xlabel("Type de formation")
plt.ylabel("Pourcentage (%)")
plt.tight_layout()
plt.show()
```

Globalement on observe une surreprésensation des femmes dans les formations par rapport aux hommes.  
Plus particulièrement dans les EFTS, IFSI, Licences, Licences LAS, PASS et les autres formations.  
Les EFTS et IFSI sont les formations avec le plus haut taux de candidates et admises (entre 80 et 90%) ces deux formations servent à poursuivre dans des domaines "traditionnellement féminins" (les soins infirmiers et le travail social).  

Au contraire les hommes eux sont plus présents dans les BTS, BUT, CPGE, écoles d'ingénieur et écoles de commerce donc des formations professionnalisantes ou assez sélective.
